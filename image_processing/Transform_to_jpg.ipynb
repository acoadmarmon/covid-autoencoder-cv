```python
import PIL.Image as pil_image
import io
import numpy as np
from PIL import Image
import os
```

```python
yourpath = os.getcwd()
for root, dirs, files in os.walk(yourpath, topdown=False):
    for name in files:
#         print(os.path.join(root, name))
#         print(os.path.splitext(os.path.join(root, name))[1].lower())
        if os.path.splitext(os.path.join(root, name))[1].lower() == ".tif":
#             if os.path.isfile(os.path.splitext(os.path.join(root, name))[0] + ".jpg"):
#                 print("A jpeg file already exists for %s" % name)
            # If a jpeg is *NOT* present, create one from the tiff.
#             else:
            if not os.path.exists(os.path.split(root)[0] + "_jpgs/" +os.path.split(root)[1]):
                os.makedirs(os.path.split(root)[0] + "_jpgs/" + os.path.split(root)[1])
#             print(os.path.split(root)[0] + "/jpgs/" + os.path.split(root)[1] + "/" + name)
#             print(name)
            outfile = os.path.splitext(os.path.join(os.path.split(root)[0] + "_jpgs/" + os.path.split(root)[1] + "/" + name))[0] + ".jpg"
            try:
#                     print("made it to try block")
                with open(os.path.join(root, name), 'rb') as f:
                    tif = pil_image.open(io.BytesIO(f.read()))
                array=np.array(tif)
                max_val=np.amax(array)
#                 print(max_val)
                normalized= (array/max_val) * 255
#                 print(np.amax(normalized))
                im = pil_image.fromarray(normalized)
                basewidth = 500
                wpercent = (basewidth/float(im.size[0]))
                hsize = int((float(im.size[1])*float(wpercent)))
                im = im.resize((basewidth,hsize), Image.ANTIALIAS)
                im = im.convert("L")
                print("Generating jpeg for %s" % name)
                im.thumbnail(im.size)
                im.save(outfile, "JPEG")
            except Exception as e:
                print(e)
```
